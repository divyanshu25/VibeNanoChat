```python
# read from imput.txt
with open("../data/input.txt", "r") as f:
    text = f.read()

# print the text
print(text[:100])
```

```python
import tiktoken

enc = tiktoken.get_encoding("gpt2")
tokens = enc.encode(text)
print(tokens[:100])

decoded = enc.decode(tokens)
print(decoded[:100])
```
